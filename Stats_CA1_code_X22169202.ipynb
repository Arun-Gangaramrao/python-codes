```python
#function for DIAGNOSTICS OF GAUSS MARKOV ASSUMPTIONS
def modelSummaryPlot(x,y,y_pred,y_test):
    # create a 2x2 grid of subplots
    fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
    residuals = y_pred - y_test
    # plotting the first graph (Residual vs Fitted)
    axs[0,0].scatter(y_pred, residuals, alpha=0.5)
    axs[0,0].set_xlabel('Fitted values')
    axs[0,0].set_ylabel('Residuals')
    axs[0,0].set_title('Residual vs. Fitted')

    # plotting the second graph (Normal QQ plot)
    sm.qqplot(residuals, line='s', ax=axs[0,1])
    axs[0,1].set_title('Normal QQ plot')

    # plotting the third graph (Scale-Location plot)
    sns.regplot(x=y_pred, y=residuals, lowess=True, line_kws={'color': 'red'}, ax=axs[1,0])
    axs[1,0].set_title('Scale-Location Plot')
    axs[1,0].set_xlabel('Fitted values')
    axs[1,0].set_ylabel('Residuals (sqrt(abs(residuals)))')

    # plotting the fourth graph (Residual vs Leverage)
    sm_model = sm.OLS(y, sm.add_constant(x))
    sm_results = sm_model.fit()
    resid = sm_results.resid
    leverage = sm_results.get_influence().hat_matrix_diag

    axs[1,1].scatter(leverage, resid, alpha=0.5)
    axs[1,1].set_xlabel("Leverage")
    axs[1,1].set_ylabel("Residuals")
    axs[1,1].set_title('Residual vs. Leverage')

    # adjust subplot spacing
    plt.tight_layout()

    # display the plot
    plt.show()
    # performing Durbin-Watson test for independence of errors
    dw = sm.stats.stattools.durbin_watson(residuals)
    print('Durbin-Watson statistic:', dw)
```

```python
#Model 1 Fig 7.1
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
x = data1.drop(['deathRate','County'], axis=1)  # Predictor variables
y = data1['deathRate']        # Target variable

#splitting the data into training and testing data
x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

#model 
ml = LinearRegression()
model=ml.fit(x_train, y_train)
#predicting values
y_pred=ml.predict(x_test)

# Evaluating the model
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Print the evaluation metrics
print("R-squared: %.4f" % r2)
print("Mean Absolute Error: %.4f" % mae)
print("Mean Squared Error: %.4f" % mse)
print("Root Mean Squared Error: %.4f" % rmse)
modelSummaryPlot(x,y,y_pred,y_test) 
```

```python
#Fig
import matplotlib.pyplot as plt
import numpy as np

# scatter plot of actual vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5, color='blue')
plt.xlabel("Actual Death Rate")
plt.ylabel("Predicted Death Rate")
plt.title("Actual vs Predicted Death Rates")
plt.grid(True)
plt.show()

# residual plot
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, y_test - y_pred, alpha=0.5, color='red')
plt.plot([min(y_pred), max(y_pred)], [0, 0], linestyle='--', color='black', linewidth=1)
plt.xlabel("Predicted Death Rate")
plt.ylabel("Residual")
plt.title("Residual Plot")
plt.grid(True)
plt.show()
```

```python
#Fig 8.2
from sklearn.preprocessing import StandardScaler

numerical_cols = [ 'Population','incidenceRate','medIncome','povertyPercent','MedianAge','MedianAgeMale','MedianAgeFemale','AvgHouseholdSize','PctMarriedHouseholds','PctNoHS18_24','PctHS18_24','PctBachDeg18_24','PctHS25_Over','PctBachDeg25_Over','PctUnemployed16_Over','PctPrivateCoverage','PctEmpPrivCoverage','PctPublicCoverage','PctPublicCoverageAlone','PctWhite','PctBlack','PctAsian','PctOtherRace'
]
# Creating a StandardScaler object
scaler = StandardScaler()

# Normalizing the data
data1[numerical_cols] = scaler.fit_transform(data1[numerical_cols])

#removing outliers
# Calculating the first and third quartiles
Q1 = data1[numerical_cols].quantile(0.25)
Q3 = data1[numerical_cols].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Defining the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Removing the rows with outliers
data = data1[~((data1[numerical_cols] < lower_bound) | (data1[numerical_cols] > upper_bound)).any(axis=1)]
data
```
